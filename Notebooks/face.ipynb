```python
import os
import pandas as pd
import numpy as np

from tsfresh import extract_features
from tsfresh.feature_extraction import MinimalFCParameters
```

```python
input_dir_path = '../analysis/ready_group/face'
output_dir = '../extracted_features/computed_ready/face'
```

```python
for filename in os.listdir(input_dir_path):
    print(filename)
    if filename == '.ipynb_checkpoints':
        continue  # Skip the .ipynb_checkpoints directory

    df_id = filename.split('_')[0]
    input_df = pd.read_csv(os.path.join(input_dir_path, filename))
    input_df = input_df.apply(pd.to_numeric, errors='coerce')
    input_df.fillna(0, inplace=True)

    input_df['window'] = input_df['window'].astype(int)

    # Increment 'window' column by 1
    input_df['window'] = input_df['window'] + 1


    settings = MinimalFCParameters()

    neutral = extract_features(input_df, column_id='neutral', column_sort='window', default_fc_parameters=settings)
    neutral.dropna(axis=1, how='all', inplace=True)

    happy = extract_features(input_df, column_id='happy', column_sort='window', default_fc_parameters=settings)
    happy.dropna(axis=1, how='all', inplace=True)

    sad = extract_features(input_df, column_id='sad', column_sort='window', default_fc_parameters=settings)
    sad.dropna(axis=1, how='all', inplace=True)

    angry = extract_features(input_df, column_id='angry', column_sort='window', default_fc_parameters=settings)
    angry.dropna(axis=1, how='all', inplace=True)
    # 
    fearful = extract_features(input_df, column_id='fearful', column_sort='window', default_fc_parameters=settings)
    fearful.dropna(axis=1, how='all', inplace=True)

    disguisted = extract_features(input_df, column_id='disgusted', column_sort='window', default_fc_parameters=settings)
    disguisted.dropna(axis=1, how='all', inplace=True)

    surprised = extract_features(input_df, column_id='surprised', column_sort='window', default_fc_parameters=settings)
    surprised.dropna(axis=1, how='all', inplace=True)

    neutral_change = extract_features(input_df, column_id='neutral_change', column_sort='window', default_fc_parameters=settings)
    neutral_change.dropna(axis=1, how='all', inplace=True)

    happy_change = extract_features(input_df, column_id='happy_change', column_sort='window', default_fc_parameters=settings)
    happy_change.dropna(axis=1, how='all', inplace=True)

    sad_change = extract_features(input_df, column_id='sad_change', column_sort='window', default_fc_parameters=settings)
    sad_change.dropna(axis=1, how='all', inplace=True)

    angry_change = extract_features(input_df, column_id='angry_change', column_sort='window', default_fc_parameters=settings)
    angry_change.dropna(axis=1, how='all', inplace=True)

    fearful_change = extract_features(input_df, column_id='fearful_change', column_sort='window', default_fc_parameters=settings)
    fearful_change.dropna(axis=1, how='all', inplace=True)

    disguisted_change = extract_features(input_df, column_id='disgusted_change', column_sort='window', default_fc_parameters=settings)
    disguisted_change.dropna(axis=1, how='all', inplace=True)
    
    surprised_change = extract_features(input_df, column_id='surprised_change', column_sort='window', default_fc_parameters=settings)
    surprised_change.dropna(axis=1, how='all', inplace=True)

    df1 = neutral.reset_index(drop=True)
    df2 = happy.reset_index(drop=True)
    df3 = sad.reset_index(drop=True)
    df4 = angry.reset_index(drop=True)
    df5 = fearful.reset_index(drop=True)
    df6 = disguisted.reset_index(drop=True)
    df7 = surprised.reset_index(drop=True)
    df8 = neutral_change.reset_index(drop=True)
    df9 = happy_change.reset_index(drop=True)
    df10 = sad_change.reset_index(drop=True)
    df11 = angry_change.reset_index(drop=True)
    df12 = fearful_change.reset_index(drop=True)
    df13 = disguisted_change.reset_index(drop=True)
    df14 = surprised_change.reset_index(drop=True)
    
    main_df = input_df.reset_index(drop=True)

    extracted_face_feat = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14], axis=1)
    face_features = pd.concat([main_df, extracted_face_feat], axis=1)
    

    #cols_to_drop = []
    #for col in cog_df_features.columns:
    #    if cog_df_features[col].nunique() == 1:
    #        cols_to_drop.append(col)

    #cog_df_features.drop(cols_to_drop, axis=1, inplace=True)

    output_filename = f"{df_id}_face_features.csv"
    face_features.to_csv(os.path.join(output_dir, output_filename), index=False)
```
