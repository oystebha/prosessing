```python
import os
import pandas as pd
import numpy as np
from tsfresh.utilities.dataframe_functions import impute
```

```python
INPUT_PATH = '../analysis/ready_group/width_wise/All/'
OUTPUT_PATH = '../analysis/ready_group/width_wise/All/all/'
```

```python
dataframes = []  # list to hold all dataframes

for filename in os.listdir(INPUT_PATH):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(INPUT_PATH, filename))
        print(f"{filename}: ",len(df.columns))
        df = df[df.window != 0]
        dataframes.append(df)  # add the dataframe to the list

result = pd.concat(dataframes, axis=0)
display(result)
output_filename= f"final.csv"


result.to_csv(os.path.join(OUTPUT_PATH, output_filename), index=False)
```
