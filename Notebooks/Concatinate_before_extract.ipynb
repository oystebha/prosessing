```python
import os
import pandas as pd
import numpy as np

from tsfresh import extract_features
from tsfresh.feature_extraction import MinimalFCParameters
```

```python
IN_PATH = "../analysis/ready_group/width_wise/10006/"
```

```python
dataframes = []  # list to hold all dataframes

for filename in os.listdir(IN_PATH):
    print(filename)
    if filename.startswith('.ipynb'):
        continue  # Skip the .ipynb_checkpoints directory
    df_id = filename.split('_')[0]
    if filename.endswith('.csv'):  # if the file is a CSV file
        df = pd.read_csv(os.path.join(IN_PATH, filename)) 
        if 'ID' in df.columns:
            df.drop("ID", inplace=True, axis=1)
        if 'success' in df.columns:
            df.drop("success", inplace=True, axis=1)
        if 'success_categorical' in df.columns:
            df.drop("success_categorical", inplace=True, axis=1)
        dataframes.append(df)

result = pd.concat(dataframes, axis=1)
result.drop("Exp_ID", inplace=True, axis=1)
result.insert(1,"ID", pd.Series([df_id]*len(result)), True)
result.insert(2,"success", pd.Series([3]*len(result)), True)
result.insert(3,"success_categorical", pd.Series(["low"]*len(result)), True)
result.columns = result.columns.map(lambda x: 'user_' + x[2] + x[3:] if x[:3].isdigit() else x)


result.fillna(0, inplace=True)

display(result)
output_filename= f"{df_id}_ww_final.csv"
result.to_csv(os.path.join(IN_PATH, output_filename), index=False)
```
